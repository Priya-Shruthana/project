# Univariate Analysis

```python
import pandas as pd
df=pd.read_csv("irisdataset.csv")
df
```

```python
df.describe()
```

```python
df["species"].value_counts().plot.bar()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x=df.index,y=df.sepal_width,hue=df.species)
```

```python
plt.plot(df.index,df.petal_length)
```

```python
plt.hist(df.petal_width)
```

```python
sns.boxplot(df.sepal_width)
```

```python
plt.pie(df.species).value_counts()
```

```python
sns.countplot(x=df["species"])
```

```python
df.head(2)
```

```python
import numpy as np
versi=df.loc[df.species=="Iris-versicolor"]
```

```python
plt.plot(versi["sepal_length"],np.zeros_like(versi["sepal_length"]))
plt.show()
```

```python
#MULTIVARIATE ANALYSIS
sns.pairplot(df,hue="species",size=3)
```
